```python
import xlsxwriter
import mysql.connector as server
import json

moneda = "MXN"
todas_emisoras = {}

db = server.connect(host="192.168.100.200", user="mensajero", passwd="password", db="db_trading")
query = db.cursor()
lista_periodo_turnos = query.execute(f"SELECT emisora, periodo FROM fundamentales")
lista_periodo_turnos = query.fetchall()

query = db.cursor(dictionary=True)
query.execute("SELECT * FROM emisoras")
todas_emisoras_bruto = query.fetchall()
db.close()

for x in todas_emisoras_bruto:
    todas_emisoras[x["emisora"]] = x["razon_social"]

for turno in lista_periodo_turnos:
    clave_pizarra_turno = turno[0]
    periodo_turno = turno[1]
    nombre_emisora = todas_emisoras[clave_pizarra_turno]
    ronda = 0

    db = server.connect(host="192.168.100.200", user="mensajero", passwd="password", db="db_trading")
    query = db.cursor()
    resultado = query.execute(f"SELECT eeff FROM fundamentales WHERE periodo = '{periodo_turno}' AND emisora = '{clave_pizarra_turno}'")
    resultado = query.fetchall()
    resultado = json.loads(resultado[0][0])
    db.close()


    """Inicio de maquila Excel"""
    workbook = xlsxwriter.Workbook(f"{clave_pizarra_turno}_{periodo_turno}_EstadosFinancieros.xlsx")
    worksheet_balance = workbook.add_worksheet('Estado de Posicion Financiera')
    worksheet_resultados_ytd = workbook.add_worksheet('Estado de Resultados YTD')
    worksheet_resultados_trimestre = workbook.add_worksheet('Estado de Resultados Trimestre')
    worksheet_flujo = workbook.add_worksheet('Estado de Flujos de Efectivo')
    negritas_centrado = workbook.add_format({'bold':True})
    centrado = negritas_centrado.set_align('center')
    cashbaby = workbook.add_format({'num_format':'#,##0.00;[Red](#,##0.00)'})
    
    for ronda in range(2):
    
        if ronda == 0:
            columna_numeros = 1
        elif ronda == 1:
            columna_numeros = 2

        #Balance General
        worksheet_balance.write(0,0,nombre_emisora,negritas_centrado)
        worksheet_balance.write(0,1,f'Cifras en {moneda}',centrado)
        worksheet_balance.write(2,columna_numeros,str(list(resultado["posicion"].keys())[0]),negritas_centrado)

        for i, (llave, valor) in enumerate(resultado["posicion"][list(resultado["posicion"].keys())[0]].items()):
            worksheet_balance.write(2+i,0,llave)
            try:
                worksheet_balance.write(2+i,columna_numeros,valor[1],cashbaby)
            except IndexError:
                pass

        #Estado de Resultados
        worksheet_resultados_ytd.write(0,0,nombre_emisora, negritas_centrado)
        worksheet_resultados_ytd.write(0,1,f'Cifras en {moneda}',centrado)
        worksheet_resultados_ytd.write(2,columna_numeros,str(list(resultado["resultado_acumulado"].keys())[0]),negritas_centrado)

        for i, (llave, valor) in enumerate(resultado["resultado_acumulado"][list(resultado["resultado_acumulado"].keys())[0]].items()):
            worksheet_resultados_ytd.write(2+i,0,llave)
            try:
                worksheet_resultados_ytd.write(2+i,columna_numeros,valor[1],cashbaby)
            except IndexError:
                pass

        worksheet_resultados_trimestre.write(0,0,nombre_emisora, negritas_centrado)
        worksheet_resultados_trimestre.write(0,1,f'Cifras en {moneda}',centrado)
        worksheet_resultados_trimestre.write(2,columna_numeros,str(list(resultado["resultado_trimestre"].keys())[0]),negritas_centrado)

        for i, (llave, valor) in enumerate(resultado["resultado_trimestre"][list(resultado["resultado_trimestre"].keys())[0]].items()):
            worksheet_resultados_trimestre.write(2+i,0,llave)
            try:
                worksheet_resultados_trimestre.write(2+i,columna_numeros,valor[1],cashbaby)
            except IndexError:
                pass

        #Flujo de Efectivo
        worksheet_flujo.write(0,0,nombre_emisora, negritas_centrado)
        worksheet_flujo.write(0,1,f'Cifras en {moneda}',centrado)
        worksheet_flujo.write(2,columna_numeros,str(list(resultado["flujos"].keys())[0]),negritas_centrado)

        for i, (llave, valor) in enumerate(resultado["flujos"][list(resultado["flujos"].keys())[0]].items()):
            worksheet_flujo.write(2+i,0,llave)
            try:
                worksheet_flujo.write(2+i,columna_numeros,valor[1],cashbaby)
            except IndexError:
                pass

        worksheet_balance.set_column('A:A',116)
        worksheet_balance.set_column('B:C',18)
        worksheet_resultados_ytd.set_column('A:A',68)
        worksheet_resultados_ytd.set_column('B:C',25)
        worksheet_resultados_trimestre.set_column('A:A',68)
        worksheet_resultados_trimestre.set_column('B:C',25)
        worksheet_flujo.set_column('A:A',108)
        worksheet_flujo.set_column('B:C',25)

    workbook.close()
    print(f"Procesado: {clave_pizarra_turno} {periodo_turno}")
```
